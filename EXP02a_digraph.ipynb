```python
def rank_comp(D):
    n=D.order()
    Q=D.incidence_matrix()
    show(Q)
    r=Q.rank()
    C=D.connected_components()
    print("The Components are ")
    show(C)
    k=len(C)
    print("The Number of Components are",k)
    if r==n-k:
        print("The Rank-Component Theorem is Verified")
    else:
        print("Something Went Wrong In The Code")
```

```python
d1=DiGraph({1:[2,8],2:[3],4:[3],5:[2,4],6:[2,4],7:[4,8],8:[2],9:[10,11],10:[11]})
show(d1)
rank_comp(d1)
d11=d1.to_undirected()
show(d11)
```

```python
d2=DiGraph({1:[2,3],2:[3],4:[1],5:[6,8],6:[8],7:[8],9:[12,11,10],10:[11,9],11:[12]})
show(d2)
rank_comp(d2)
d22=d2.to_undirected()
show(d22)
```
