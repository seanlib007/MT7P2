```python
def dis(G,i,j):
    s1=G.all_paths(i,j)
    s11=[len(i)-1 for i in s1]
    k=min(s11)
    print("the distance between",i,"and",j,"is",k)

dis(T,2,11)
dis(S,1,3)
dis(BS,3,8)
dis(P,2,5)
```

```python
def dis1(G,i,j):
    s1=G.all_paths(i,j)
    s11=[len(i)-1 for i in s1]
    k=min(s11)
    return k
def rec(G):
    ver=G.vertices()
    rad=G.order()-1
    for i in ver:
        L=[dis(G,i,j) for j in ver]
        print("the eccentricity of the vertex",i,"is",max(L))
        if max(L)<rad:
            rad=max(L)
    print("THE radius of the graph is ",rad)
    cen=[i for i in ver if max([dis(G,i,j)for j in ver])==rad]
    print("the center of the graph is",cen,"\n\n")
```

```python
rec(S)
rec(BS)
rec(T)
rec(P)
```

```python
def dis2(G,i,j):
    s1=G.all_paths(i,j)
    s11=[len(i)-1 for i in s1]
    k=min(s11)
    return k
def rec1(G):
    ver=G.vertices()
    rad=G.order()-1
    for i in ver:
        L=[dis2(G,i,j) for j in ver]
        if max(L)<rad:
            rad=max(L)
    cen=[i for i in ver if max([dis2(G,i,j)for j in ver])==rad]
    return(rad,cen)
def cen_adj(T):
    if T.is_tree():
        cen=rec1(T)[1]
        if len(cen)<3:
            print("there are atmost 2 center of the tree")
        if len(cen)==1:
            print("there is 1 center of the tree\n\n")
        if len(cen)==2:
            if(cen[0],cen[1])in T.edges(labels=false):
                print("the 2 centres are adjacent\n\n")
```

```python
cen_adj(S)
cen_adj(BS)
cen_adj(T)
cen_adj(P)
```

```python
def spantree(g):
    c=g.spanning_trees_count()
    l=g.laplacian_matrix()
    s=l.delete_rows([1]).delete_columns([1]).det()
    if c==s:
        print("no. of spanning trees is equal to cofactor")
    else:
        print("no. of spanning trees is not equal to cofactor")
spantree(S)
spantree(BS)
spantree(T)
spantree(P)
```
